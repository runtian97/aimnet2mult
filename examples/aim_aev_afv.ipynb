```python
import torch
from openbabel import pybel
```

```python

filename = '/Users/nickgao/Desktop/pythonProject/aimnet2mult/examples/run/model_fid0.jpt'
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = torch.jit.load(filename, map_location=device)

mol_file = '/Users/nickgao/Desktop/pythonProject/local_code_template/test_sdfs/CAMVES_a.sdf'
mol = next(pybel.readfile('sdf', mol_file))

coord = torch.as_tensor([a.coords for a in mol.atoms]).unsqueeze(0).to(device)
numbers = torch.as_tensor([a.atomicnum for a in mol.atoms]).unsqueeze(0).to(device)
charge = torch.as_tensor([mol.charge]).to(device)
mult = torch.as_tensor([mol.spin]).to(device)

_in = dict(
    coord=coord,
    numbers=numbers,
    charge=charge,
    mult=mult,
)

_out = model(_in)

# coord
v_coord = _out['coord']

# numbers(atomic number in periodic table)
v_numbers = _out['numbers']

# charge(total charge of the mol)
v_charge = _out['charge']

# pad_mask
# v_pad_mask = _out['pad_mask']

# _natom (total number of atoms )
v_natom = _out['_natom']

#*********** below are predicted values*********#

# a (atomic feature)
v_a = _out['a']

# d_ij (distance vector: distance between atoms)
v_d_ij = _out['d_ij']

# u_ij (unit distance vector)
# v_u_ij = _out['u_ij']

# gs
v_gs = _out['gs']

# gv
v_gv = _out['gv']

# charges(predicted for each atom)
v_charges = _out['charges']

# energy
v_energy = _out['energy']

# *********most important part about the algorithm
# AIMNet2 type convolution: encoding of local environment which combines geometry of local environment and atomic features.

# Atomic feature vector
afv = _out['a'].flatten(-2, -1)
# AIM layer
aim = _out['aim']

```
